```python
import numpy as np
import cv2
import os
from google.colab.patches import cv2_imshow
```

```python
name=input("please enter the name of image with extension")
IMAGE_PATH = "INPUT_IMAGES"
IMAGE_PATH =os.path.join(IMAGE_PATH,name)
```

```python
img = cv2.imread(IMAGE_PATH)

image = cv2.imread(IMAGE_PATH)
image= cv2.GaussianBlur(image, (155, 153), 0)
h,w,c=image.shape
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

mask1 = cv2.inRange(hsv, (0,90,50), (4,255,255))
mask2 = cv2.inRange(hsv, (175,90,50), (180,255,255))
red_mask = cv2.bitwise_or(mask1, mask2 )


green_lower = np.array([36, 25, 25], np.uint8)
green_upper = np.array([80, 255, 255], np.uint8)
green_mask = cv2.inRange(hsv, green_lower, green_upper)


blue_lower = np.array([94, 80, 2], np.uint8)
blue_upper = np.array([120, 255, 255], np.uint8)
blue_mask = cv2.inRange(hsv, blue_lower, blue_upper)


yellow_lower = np.array([25, 50, 70], dtype="uint8")
yellow_upper = np.array([35, 255, 255], dtype="uint8")
yellow_mask = cv2.inRange(hsv, yellow_lower, yellow_upper)

orange_mask = cv2.inRange(hsv,(6, 90, 90), (24, 255, 255) )

purple_mask = cv2.inRange(hsv,(129, 70, 50), (158, 255, 255) )

black_mask=cv2.inRange(hsv,(0, 0, 0), (180, 255, 40) )

pink_mask=cv2.inRange(hsv,(130, 20, 220), (170, 255, 255) )

kernal = np.ones((5, 5), "uint8")

red_mask = cv2.dilate(red_mask, kernal)
res_red = cv2.bitwise_and(image, image, 
                          mask = red_mask)

black_mask = cv2.dilate(black_mask, kernal)
res_black = cv2.bitwise_and(image, image, 
                          mask = black_mask)

green_mask = cv2.dilate(green_mask, kernal)
res_green = cv2.bitwise_and(image, image,
                            mask = green_mask)

blue_mask = cv2.dilate(blue_mask, kernal)
res_blue = cv2.bitwise_and(image, image,
                           mask = blue_mask)

yellow_mask = cv2.dilate(yellow_mask, kernal)
res_yellow = cv2.bitwise_and(image, image,
                           mask = yellow_mask)


orange_mask = cv2.dilate(orange_mask, kernal)
res_orange = cv2.bitwise_and(image, image,
                           mask = orange_mask)

purple_mask = cv2.dilate(purple_mask, kernal)
res_pink = cv2.bitwise_and(image, image,
                            mask = purple_mask)

pink_mask = cv2.dilate(pink_mask, kernal)
res_pink = cv2.bitwise_and(image, image,
                            mask = pink_mask)

contours, hierarchy = cv2.findContours(red_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)
ratios=h*w/18
for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
      x, y, w, h = cv2.boundingRect(contour)
      image = cv2.rectangle(img, (x, y), 
                                   (x + w, y + h), 
                                   (0, 0, 255), 3)

      cv2.putText(img, "Red", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (0, 0, 255),2)    
        
contours, hierarchy = cv2.findContours(black_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)

for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y), 
                                   (x + w, y + h), 
                                   (0, 0, 0), 3)

        cv2.putText(img, "Black", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (0, 0, 0),2)   


contours, hierarchy = cv2.findContours(green_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)

for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y), 
                                   (x + w, y + h),
                                   (0, 255, 0), 3)

        cv2.putText(img, "Green", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX, 
                    0.8, (0, 255, 0),2)


contours, hierarchy = cv2.findContours(blue_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)
for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y),
                                   (x + w, y + h),
                                   (255, 0, 0), 3)

        cv2.putText(img, "Blue", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (255, 0, 0),2)

        

        
contours, hierarchy = cv2.findContours(yellow_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)
for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y),
                                   (x + w, y + h),
                                   (0, 255, 255), 3)

        cv2.putText(img, "yellow", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (0, 255, 255),2)        


        
        


contours, hierarchy = cv2.findContours(orange_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)

for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y), 
                                   (x + w, y + h), 
                                   (0, 165, 255), 3)

        cv2.putText(img, "orange", (x+15, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (0, 165, 255),2)  
       
        
        
        

contours, hierarchy = cv2.findContours(purple_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)

for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y), 
                                   (x + w, y + h), 
                                   (217,156,177), 3)

        cv2.putText(img, "purple", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (197,136,157),2) 



contours, hierarchy = cv2.findContours(pink_mask,
                                       cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)

for pic, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    if(area > ratios):
        x, y, w, h = cv2.boundingRect(contour)
        image = cv2.rectangle(img, (x, y), 
                                   (x + w, y + h), 
                                   (106, 51, 170), 3)

        cv2.putText(img, "pink", (x, y+15),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (106, 51, 170),2)  






from google.colab.patches import cv2_imshow

cv2_imshow(image)
OUTPUT_PATH=os.path.join("Result","marked"+name)
cv2.imwrite(OUTPUT_PATH,image)
cv2.waitKey(0)
```
